```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid') 
```

```python
df = pd.read_csv('train.csv')
df.head() 
```

```python
print("Shape of dataset:", df.shape)
df.info() 
```

```python
df.describe()
df.isnull().sum() 
```

```python
# Missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())
# Missing Embarked with most common port
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
# Drop Cabin column
df.drop('Cabin', axis=1, inplace=True) 
```

## Data Cleaning
In this step, I filled missing Age values with the median age and filled missing Embarked values with the most common port.  
I also dropped the Cabin column because it had too many missing values.

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8,6))
sns.histplot(df['Fare'], bins=20, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
